```python
import json
import re

def convert_to_nlp_data(dump_files, output_file, label):
    f = open(output_file, 'w')

    idx = 0
    for dump_file in dump_files:
        file=open(dump_file)
        data = file.read()
        jsonData = json.loads(data)
#         print(jsonData.keys())

        for key in jsonData.keys():
            if key != 'created':
                thread = jsonData[key]
                stack = '\n'.join(thread['stack'])
                file_st = open('dumps/threads/{0:07d}.txt'.format(idx), 'w')
                idx = idx + 1
                file_st.write(stack)
                file_st.close()
                # thread의 이름에서 영어, 띄워쓰기만 남기고 나머지(숫자, 기호 등) 삭제 
                name = re.sub('[^ a-zA-Z]','',thread['name'])
                stack = re.sub('^\"[^\"]*\"', name, stack)
                
                #스레드의 고유번호(#1234), 포인터 저장 위치(형식 [0x000], <0x000>) 삭제
                stack = re.sub('#[0-9]+|\[[^)]*?\]|\<[^)]*?\>', '', stack)
                
                #공백, .(온점), (, ), ,(반점) 기준으로 split
                words = re.split(' |\.|\(|\)|\n|,', stack)
                
                dehydrated_words = []
                for word in words:
                    #비워있는 word 스킵
                    if word == '' : continue
                    
                    #java: 형식으로 되어있는 코드 행 삭제, prio=, os_prio, tid, nid 가 포함된 단어 삭제
                    #javax, at, java, com, org 처럼 의미없이 자주 출현하는 단어 삭제(불용어 처리)
                    if word.find('java:') == -1 and word.find('prio=') == -1 and word.find('id=') == -1 and \
                    word != 'javax' and word != 'at' and word != 'java' and word != 'com' and word != 'org':
                        dehydrated_words.append(word)
                        print(word)
                if not label:
                    f.write(' '.join(dehydrated_words) + '\n')
                else:
                    if stack.find('at com.mysql.cj.protocol.FullReadInputStream.readFully') != -1:
                    print(dehydrated_words)
                    f.write(' '.join(dehydrated_words) + ',hang\n')
                  # elif stack.find('at java.security.SecureRandom.generateSeed(SecureRandom.java') != -1:
                  #   print(hydrated_words)
                  #   f.write(' '.join(dehydrated_words) + ', secure_random\n')
                  # elif stack.find('java.net.SocketOutputStream.socketWrite0') != -1:
                  #   print(hydrated_words)
                  #   f.write(' '.join(dehydrated_words) + ', socket_write\n')
                    elif stack.find('a org.apache.tomcat.util.log.SystemLogHandler') != -1:
                        print(dehydrated_words)
                        f.write(' '.join(dehydrated_words) + ',system_log_lock\n')
                    elif stack.find('java.lang.Thread.State: BLOCKED (on object monitor)') != -1:
                        print(dehydrated_words)
                        f.write(' '.join(dehydrated_words) + ',thread_block\n')
                    elif stack.find('at org.apache.http.impl.DefaultBHttpClientConnection.receiveResponseHeader') != -1:
                        print(dehydrated_words)
                        f.write(' '.join(dehydrated_words) + ',http_client_slow\n')
                    elif stack.find('at org.infinispan.affinity.impl.KeyAffinityServiceImpl.getKeyForAddress') != -1:
                        print(dehydrated_words)
                        f.write(' '.join(dehydrated_words) + ',infinispan_lock\n')
                    elif stack.find('at sun.net.www.protocol.http.HttpURLConnection.getNewHttpClient') != -1:
                        print(dehydrated_words)
                        f.write(' '.join(dehydrated_words) + ',url_conn_slow\n')
                    elif stack.find('at sun.nio.ch.EPollArrayWrapper.epollWait') != -1:
                        print(dehydrated_words)
                        f.write(' '.join(dehydrated_words) + ',url_conn_slow\n')
                    elif stack.find('at oracle.jdbc.driver.T4CSocketInputStreamWrapper.readNextPacket') != -1:
                        print(dehydrated_words)
                        f.write(' '.join(dehydrated_words) + ',oracle_slow\n')
                    elif stack.find('at java.util.concurrent.Semaphore.tryAcquire') != -1:
                        print(dehydrated_words)
                        f.write(' '.join(dehydrated_words) + ',semaphore_lock\n')
                    elif stack.find('at org.apache.commons.pool.impl.GenericObjectPool.borrowObject') != -1:
                        print(dehydrated_words)
                        f.write(' '.join(dehydrated_words) + ',db_pool_wait\n')
                    elif stack.find('org.jboss.as.clustering.web.infinispan.DistributedCacheManager.acquireSessionOwnership') != -1:
                        print(dehydrated_words)
                        f.write(' '.join(dehydrated_words) + ',infinispan_ownership\n')
                    else:
                        f.write(' '.join(dehydrated_words) + ',normal\n')

    f.close()
```
